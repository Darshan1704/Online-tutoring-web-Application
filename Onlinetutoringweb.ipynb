```python
import streamlit as st
import pandas as pd

# Hardcoded username and password for demonstration purposes
demo_username_teacher = "tted"
demo_password_teacher = "qwer12"
demo_username_student = "stu123"
demo_password_student = "123456"

# Mock class schedule database
class_schedule_db = pd.DataFrame(columns=["ClassID", "Subject", "Tutor", "Time"])

# Mock attendance database
attendance_db = pd.DataFrame(columns=["ClassID", "StudentID", "Attendance"])

# Function to get today's timetable
def get_timetable(user_type, user_id):
    # Dummy timetable for illustration
    if user_type == "teacher":
        # Replace this with the actual timetable retrieval logic for teachers
        timetable = {"Monday": "Math Class", "Tuesday": "English Class", "Wednesday": "Science Class"}
    else:
        # Replace this with the actual timetable retrieval logic for students
        timetable = {"Monday": "History Class", "Tuesday": "Math Class", "Wednesday": "Physics Class"}

    return timetable

# Streamlit app
def main():
    st.title("Online Tutoring System")

    # Initialize session state
    user_type = st.session_state.get("user_type", None)
    username = st.session_state.get("username", None)

    # Check if the user is already logged in
    if user_type is None or username is None:
        # Login functionality
        user_type = st.radio("Login as", ("Teacher", "Student"))
        username = st.text_input("Username:")
        password = st.text_input("Password:", type="password")

        if st.button("Login"):
            if authenticate_user(user_type.lower(), username, password):
                st.session_state.user_type = user_type.lower()
                st.session_state.username = username
                st.success(f"Logged in as {user_type} - {username}")
                show_dashboard(user_type, username)
            else:
                st.error("Invalid credentials. Please try again.")
    else:
        # User is already logged in, skip login and show the dashboard directly
        st.success(f"Logged in as {user_type.capitalize()} - {username}")
        show_dashboard(user_type, username)

def show_dashboard(user_type, username):
    st.subheader(f"Welcome, {user_type} - {username}!")

    # Class Scheduler
    st.header("Class Scheduler")
    schedule_classes()

    # Attendance Capturing
    st.header("Attendance Capturing")
    capture_attendance()

    # Document Upload
    st.header("Document Upload")
    document_upload()

    # NLP Chatbot
    st.header("NLP Chatbot")
    nlp_chatbot(user_type)

def authenticate_user(user_type, username, password):
    if user_type == "teacher" and username == demo_username_teacher and password == demo_password_teacher:
        return True
    elif user_type == "student" and username == demo_username_student and password == demo_password_student:
        return True
    else:
        return False

def schedule_classes():
    st.subheader("Schedule a Class")

    subject = st.text_input("Subject:")
    tutor = st.selectbox("Select Tutor:", list(["Tutor1", "Tutor2"]))  # Add your tutors here
    time = st.text_input("Time (e.g., 2023-01-01 10:00 AM):")

    if st.button("Schedule Class"):
        class_id = len(class_schedule_db) + 1
        class_schedule_db.loc[len(class_schedule_db)] = [class_id, subject, tutor, time]
        st.success(f"Class scheduled successfully with ID: {class_id}")

    st.dataframe(class_schedule_db)

def capture_attendance():
    st.subheader("Capture Attendance")

    class_id = st.selectbox("Select Class:", class_schedule_db["ClassID"])
    student_id = st.text_input("Student ID:")
    attendance_status = st.radio("Attendance Status:", ["Present", "Absent"])

    if st.button("Capture Attendance"):
        attendance_db.loc[len(attendance_db)] = [class_id, student_id, attendance_status]
        st.success("Attendance captured successfully")

    st.dataframe(attendance_db)

def document_upload():
    st.subheader("Document Upload")

    file = st.file_uploader("Upload File:", type=["pdf", "doc", "docx"])

    if file is not None:
        st.success("File uploaded successfully")

def nlp_chatbot(user_type):
    st.subheader("Chat with the NLP Chatbot")

    user_input = st.text_input("Enter your message:")
    st.write("Chatbot's response will appear here.")

    if st.button("Send"):
        response = generate_chatbot_response(user_input, user_type)
        st.write(f"Chatbot: {response}")

def generate_chatbot_response(user_input, user_type):
    # Common responses
    common_responses = {
        "greeting": "Hello! How can I assist you?",
        "unknown": "I'm sorry, I don't have information on that.",
    "What is the capital of France?": "The capital of France is Paris.",
    "Who is the president of the United States?": "The president of the United States is Joe Biden.",
    "What is the square root of 25?": "The square root of 25 is 5.",
    "What is the largest planet in our solar system?": "The largest planet in our solar system is Jupiter.",
    }

    # Special case for the president query
    if "who is the president of the united states?" in user_input.lower():
        return "The president of the United States is Joe Biden."

    # User-specific responses
    user_specific_responses = {
        "teacher": {
            "schedule": "As a teacher, you can schedule classes using the 'Schedule a Class' section.",
            "timetable": "Your timetable for today is: {}".format(get_timetable("teacher", "your_teacher_id")),
        },
        "student": {
            "attendance": "As a student, you can capture attendance using the 'Capture Attendance' section.",
            "timetable": "Your timetable for today is: {}".format(get_timetable("student", "your_student_id")),
        },
    }

    # NLP logic to determine user intent
    if "schedule" in user_input.lower():
        return user_specific_responses[user_type]["schedule"]
    elif "timetable" in user_input.lower():
        return user_specific_responses[user_type]["timetable"]
    elif any(greeting_word in user_input.lower() for greeting_word in ["hi", "hello", "hey"]):
        return common_responses["greeting"]
    else:
        return common_responses["unknown"]

if __name__ == "__main__":
    main()
```
